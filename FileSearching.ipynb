```python
# import modules
import pandas as pd
import os
import zipfile
print("import modules")

# list file paths 
years = list(range(2023, 2025))
domain = "Sierras"
WW_asoDir = r"W:/Spatial_SWE/ASO/"
SNM_asoDir = r"J:/Spatial_SWE/ASO/"
print("file paths established")

for year in years:
    aso_dir = f"{SNM_asoDir}/{year}/"
    extensions = ("snowdepth_3m.tif", "swe_50m.tif")
    
    # create an if then statement on if a year folder exists, dont create it, if not, create it, 
    folder_path = f"C:/Users/etyrr/OneDrive/Documents/CU_Grad/ASOML/{domain}/{year}/"
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)  
        print(f"Created folder: {folder_path}")
    else:
        print(f"Folder already exists: {folder_path}")
        
    # check for snow dpeth and create that folder
    sd_path = f"C:/Users/etyrr/OneDrive/Documents/CU_Grad/ASOML/{domain}/{year}/SD/"
    if not os.path.exists(sd_path):
        os.makedirs(sd_path)  
        print(f"Created folder: {sd_path}")
    else:
        print(f"Folder already exists: {sd_path}")

    # check for swe and create that folder
    swe_path = f"C:/Users/etyrr/OneDrive/Documents/CU_Grad/ASOML/{domain}/{year}/SWE/"
    if not os.path.exists(swe_path):
        os.makedirs(swe_path)  
        print(f"Created folder: {swe_path}")
    else:
        print(f"Folder already exists: {swe_path}")

    # make a list of all zip files and loops through them
    zip_files = [f for f in os.listdir(aso_dir) if f.endswith(".zip")]

    for zip_filename in zip_files:
        zip_path = os.path.join(aso_dir, zip_filename)  # Full path to ZIP file
        print(" ")
        print(f"Processing ZIP file: {zip_path}")
        
        with zipfile.ZipFile(zip_path, 'r') as zip_file:
            matching_files = [file for file in zip_file.namelist() if file.endswith(extensions)]
            print("Files ending with", extensions, ":", matching_files)

            for file in matching_files:
                    # Determine destination based on file extension
                    if file.endswith("/"):  
                        print(f"Skipping directory: {file}")
                        continue  
                    if file.endswith("snowdepth_3m.tif"):
                        extract_path = os.path.join(sd_path, os.path.basename(file))
                    elif file.endswith("swe_50m.tif"):
                        extract_path = os.path.join(swe_path, os.path.basename(file))
                    else:
                        continue  
    
                    # Extract the specific file to the correct folder
                    with zip_file.open(file) as source, open(extract_path, "wb") as target:
                        target.write(source.read())

                    print(f"Extracted {file} to {extract_path}")
    print(f"done with {year}")
    print(" ")
```
